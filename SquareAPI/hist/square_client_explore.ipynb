Motivation:
Consider the following scenario: if limit=2, (i.e., each request only return two records at most), then how can we list of the record from the SquareAPI. 
This is an important question because, the maxium record limit by SquareAPI is 100, we need to handle this limitation to scan all the record in interests.

```python
# datetime method
from datetime import datetime, timezone, timedelta
from dateutil.relativedelta import *
local_time = datetime.now(timezone.utc).astimezone()
print(local_time.isoformat())
print((local_time - timedelta(days=1)).isoformat())
print((local_time + relativedelta(months=+6)).isoformat())
```

```python
# list payment

# arg_dict ={'begin_time' : '2021-01-02T00:00:01.007Z',
# 'end_time' : '2021-06-02T00:00:01.007Z',
# 'limit' : 2}

# get the record from past 6 months
arg_dict ={'begin_time' : (local_time - relativedelta(months=+6)).isoformat(),
'end_time' : local_time.isoformat(),
'limit' : 2}

# tips: using **arg_dict to parse dict to a argument list
result = payments_api.list_payments(**arg_dict)

# notice that the result.body has a new key-value pair called cursor. 
# This is what we will rely on to request and list all the record
print(result.body.keys())
print(result.body['cursor'])

df_flat = pd.json_normalize(json.loads(result.text), record_path="payments")
df_flat.head()
df_flat[['id','created_at', 'updated_at', 'note', 'total_money.amount', 'total_money.currency']].head()
```

```python
# use a while loop to scan all the record
cursor=None

arg_dict ={'begin_time' : (local_time - relativedelta(months=+6)).isoformat(),
'end_time' : (local_time + relativedelta(months=+6)).isoformat(),
'limit' : 2}

# tips: using **arg_dict to parse dict to a argument list
result = payments_api.list_payments(**arg_dict)
result_full_text = result.body.get("payments")

while (result.body.get('cursor') is not None):
    arg_dict['cursor'] = result.body.get('cursor')
    result = payments_api.list_payments(**arg_dict)
#     print(result.body.get('cursor'))
    result_full_text += result.body.get("payments")
    

df_flat = pd.DataFrame.from_dict(result_full_text)
df_flat.head()
df_flat[['id','created_at', 'updated_at', 'note', 'total_money']].head(15)
```

```python
# to-do: while loop here is a bit ugly
# alternatives: functional programming style

def get_full_text(payments_api, arg_dict, full_text=[]):
    result = payments_api.list_payments(**arg_dict)
    cursor = result.body.get('cursor')
    if cursor is not None:
        full_text += result.body.get("payments")
        arg_dict['cursor'] = cursor
        get_full_text(payments_api, arg_dict, full_text)
    return full_text


arg_dict ={'begin_time' : (local_time - relativedelta(months=+6)).isoformat(),
'end_time' : (local_time + relativedelta(months=+6)).isoformat(),
'limit' : 5}  

list0 = get_full_text(payments_api, arg_dict)
# print(list0)

df_flat = pd.DataFrame.from_dict(list0)
df_flat.head()
```
